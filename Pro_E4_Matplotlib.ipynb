#### 1-a)

*Wienwahl-Ergebnisse $2020$ (Wahlbeteiligung $65.27%$)*: <br> 
SPÖ $41.62$%, ÖVP $20.43$%, Grüne $14.8$%, NEOS $7.47$%, FPÖ $7.11$%, HC $3.27$%, Sonstige $5.29$%.

 Erstellen Sie einen Plot mit $2$ Subplots: einem **Balkendiagramm** und einem **Kuchendiagramm** zu den Wienwahl-Daten. Die Gruppen sowie Achsen/Titel sollen dabei sinnvoll beschriftet werden und die Farben akkurat gewählt werden (diese können in einem Vektor übergeben werden, `'color' = [...]` für das **Balkendiagramm**, `'colors' = [...]` für das **Kuchendiagramm**). Beim **Kuchendiagramm** soll 'Sonstige' abgespaltet dargestellt werden.

```python
import numpy as np
import matplotlib.pyplot as plt 


plt.subplot(121)
groups = ('SPOE', 'OEVP', 'GRUENE', 'NEOS', 'FPOE' ,'HC', 'SONSTIGE')
x = np.arange(len(groups))
data = [41.62, 20.43, 14.8, 7.47, 7.11, 3.27, 5.29]
color = ['r','turquoise','g','deeppink','b','midnightblue','grey']

plt.bar(x, data, align = 'center', color = color)
plt.xticks(x, groups, rotation = 45)
plt.xlabel('(Wahlbeteiligung 65.27%)')
plt.ylabel('Prozent (%)')
plt.title('Wienwahl-Ergebnisse 2020')


plt.subplot(122)
labels = ('SPOE', 'OEVP', 'GRUENE', 'NEOS', 'FPOE' ,'HC', 'SONSTIGE')
data = [41.62, 20.43, 14.8, 7.47, 7.11, 3.27, 5.29]
explode = (0,0,0,0,0,0,0.5)
colors = ['r','turquoise','g','deeppink','b','midnightblue','grey']

plt.pie(data, labels = labels, explode = explode , autopct = '%.2f%%', shadow = True, colors = colors)
plt.title('Wienwahl-Ergebnisse 2020')
plt.figure(figsize=(10,5))

plt.show()
```

$$f(x)=x^{2}sin(x)cos(x)^{3}$$
im Intervall $[−2,2)$ für jeweils $5,10,15$ und $100$ äquidistante $x$-Werte. Erstellen Sie für jede Anzahl der $x$-Werte einen Subplot, wobei die Funktionswerte jeweils durch Punkte, die linear verbunden sind, dargestellt werden sollen. Jeder Subplot soll sich weiters durch unterschiedliche Symbole der Punkte und Farben unterscheiden. Geben Sie den gesamten Plot (mit den $4$ Subplots) aus.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(221)
x = np.linspace(-2,2,5)
y = x**2*np.sin(x)*(np.cos(x))**3
plt.plot(x,y,':sr', )


plt.subplot(222)
x = np.linspace(-2,2,10)
y = x**2*np.sin(x)*(np.cos(x))**3
plt.plot(x,y,':^g')


plt.subplot(223)
x = np.linspace(-2,2,15)
y = x**2*np.sin(x)*(np.cos(x))**3
plt.plot(x,y,':db')


plt.subplot(224)
x = np.linspace(-2,2,100)
y = x**2*np.sin(x)*(np.cos(x))**3
plt.plot(x,y,':oy',markersize =4) # markersize 디폴트값 =6



plt.rc('lines', linewidth = 0.5)
plt.figure(figsize=(10, 5)) # 그래프 크기조절
plt.show()
```

### FRAGE : Wird die oben angegebene Größe beibehalten?

```python
import numpy as np
import matplotlib.pyplot as plt

for n in range(3,7):
    N = 10**n
    bins = 100

    a = np.random.randn(N)
    b = np.random.rand(N)

    plt.subplot(121)
    plt.hist(a, bins= bins)
    plt.xlabel(f'N ={N}, bins ={bins}')

    plt.subplot(122)
    plt.hist(b, bins= bins)
    plt.xlabel(f'N = {N}, bins ={bins}')

    plt.show()
```

#### 4)

Stellen Sie $e^{c∗x}$ für $c=1,2,3,4$ und $x∈[0.1,10)$ in einem Plot mit logarithmischer y-Achsen Skalierung dar. Die Graphen sollen für jedes $c$ eine *Legende* haben und der Plot einen *Titel*. Implementieren Sie dann analog die Darstellung von $c∗x$ ohne Achsenskalierung. Erklären Sie warum die Plots gleich aussehen.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.1, 10., 0.1)
plt.yscale('log')
plt.plot(x,np.e**(1*x),'C3',label ='c=1');plt.plot(x,np.e**(2*x),'C1',label ='c=2')
plt.plot(x,np.e**(3*x),'C2',label ='c=3');plt.plot(x,np.e**(4*x),'C0',label ='c=4')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y=e^{cx}$')
plt.legend()
plt.show()
```
